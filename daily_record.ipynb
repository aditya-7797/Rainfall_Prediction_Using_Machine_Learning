```python
import pandas as pd

# Load your original dataset (make sure to update the file path)
df = pd.read_csv('final_processed_dataset.csv')

# Convert 'date_time' column to datetime and extract the date, day, and month
df['date'] = pd.to_datetime(df['date_time']).dt.date
df['day'] = pd.to_datetime(df['date_time']).dt.day
df['month'] = pd.to_datetime(df['date_time']).dt.month

# Group by 'date' to aggregate daily data
daily_aggregated_df = df.groupby(['date', 'day', 'month']).agg({
    'precipMM': 'sum',           # Sum of precipitation over 24 hours
    'humidity': 'mean',          # Mean humidity over the day
    'maxtempC': 'max',           # Maximum temperature of the day
    'mintempC': 'min',           # Minimum temperature of the day
    'DewPointC': 'mean',         # Mean DewPoint over the day
    'WindGustKmph': 'mean',      # Mean wind gust speed over the day
    'cloudcover': 'mean',        # Mean cloud cover over the day
    'pressure': 'mean',          # Mean pressure over the day
    'visibility': 'mean',        # Mean visibility over the day
    'windspeedKmph': 'mean'      # Mean wind speed over the day
}).reset_index()

# Save the aggregated dataset to a new CSV file
daily_aggregated_df.to_csv('daily_aggregated_dataset_with_day_month.csv', index=False)

print("Dataset saved successfully!")
```
